## Getting top "count" repos with topic "tag" assigned to them.

```python
def get_sample_repo_with_tag(tag,count):
    url = 'https://api.github.com/search/repositories'
    root = 'https://github.com/'
    print('getting started')
    done = False
    while(not done):
        try:
            print('trying to connect')
            response = requests.get(url+'?q=topic:'+tag+'+sort:stars', timeout=5)
            print(response.status_code)
        except:
            print('not connected')
            continue
        repos = [].copy()
        if response.status_code != 200:
            print('not 200')
            continue
        for item in response.json()['items'][:count]:
            link = root+item['full_name']
            stargazers = item['stargazers_count']
            repos.append({'link':link,
                          'stargazers':stargazers})
        done = True
    print('return top')
    return repos
```

## Set "tag" and "count" and run to see samples

```python
with open('topics.json','r') as f:
    topics = json.load(f)
l_topics = sorted(list(topics.keys()))
n = len(l_topics)
count = 20
got = 0

for tag in l_topics:
    try:
        print('trying to find')
        top = topics[tag]['top_20']
        print('fount '+tag)
        print('@@'*20+' '+tag+' done',end='\n\n')
        got += 1
        print('#'*20+' '+str(got)+' done out of '+str(n),end='\n\n')
    except KeyError:
        print('not found')
        top = get_sample_repo_with_tag(tag,count)
        topics[tag]['top_20'] = top
        print('@@'*25+' '+tag+' done\n')
        got += 1
        print('#'*20+' '+str(got)+' done out of '+str(n),end='\n\n')
        if got % 10 == 0 or got == n:
            with open('topics_backup.json','w') as f:
                json.dump(topics,f)
            print('$$$'*20+' '+str(got)+' topics saved',end='\n\n')
```
